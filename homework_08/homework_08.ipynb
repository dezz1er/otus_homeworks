```python
import numpy as np
import matplotlib.pyplot as plt
```

Сгенерируем случайную матрицу

```python
n = 100
A = np.random.randn(n, n)
```

Выполним SVD

```python
U, S, Vt = np.linalg.svd(A, full_matrices=False)
```

Построим аппроксимацию ранга

```python
r_values = np.arange(1, n + 1)
errors = []

for r in r_values:
    # Собираем аппроксимацию ранга r
    A_r = U[:, :r] @ np.diag(S[:r]) @ Vt[:r, :]
    # Считаем ошибку (норма Фробениуса)
    err = np.linalg.norm(A - A_r, ord='fro')
    errors.append(err)
```

Визуализируем ошибку

```python
plt.figure(figsize=(8, 5))
plt.plot(r_values, errors, marker='o')
plt.title("Ошибка аппроксимации матрицы в зависимости от ранга r")
plt.xlabel("Ранг r")
plt.ylabel("Ошибка E(r) = ||A - Â_r||_F")
plt.grid(True)
plt.show()
```
