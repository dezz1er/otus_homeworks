### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
import numpy as np

arr = np.zeros(10)
arr[4] = 1
arr_2d = arr.reshape(5, 2)
print(arr_2d)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
import numpy as np


arr = np.arange(10, 50)
reversed_arr = arr[::-1]
print(reversed_arr)
even_elements = arr[arr % 2 == 0]
print(even_elements)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
import numpy as np

random_array = np.random.random((4, 3, 2))

print(random_array)



array_min = random_array.min()
array_max = random_array.max()

print(f"\nМинимальное значение в массиве: {array_min}")
print(f"Максимальное значение в массиве: {array_max}")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение.

```python
import numpy as np


matrix_a = np.random.random((6, 4))
matrix_b = np.random.random((4, 3))
result_matrix = np.dot(matrix_a, matrix_b)
print(result_matrix)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
import numpy as np


random_array = np.random.random((7, 7))

array_mean = random_array.mean()
array_std = random_array.std()

normalized_array = (random_array - array_mean) / array_std
print(array_mean)
print(array_std)
print(normalized_array)
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


tips = sns.load_dataset('tips')
```

### Посмотрите на первые 5 строчек

```python
print(tips.head())
```

### Узнайте сколько всего строчек и колонок в данных

```python
print(f"Строчек: {tips.shape[0]}")
print(f"Колонок: {tips.shape[1]}")
```

### Проверьте есть ли пропуски в данных

```python
print(tips.isnull().sum())
print("Пропусков нет!" if tips.isnull().sum().sum() == 0 else "Есть пропуски!")
```

### Посмотрите на распределение числовых признаков

```python
numerical_cols = tips.select_dtypes(include=[np.number]).columns
print(tips[numerical_cols].describe())
```

### Найдите максимальное значение 'total_bill'

```python
max_bill = tips['total_bill'].max()
print(f"Максимальный total_bill: ${max_bill:.2f}")
```

### Найдите количество курящих людей

```python
smoker_counts = tips['smoker'].value_counts()
print(f"Курящих: {smoker_counts.get('Yes', 0)}")
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
avg_bill_by_day = tips.groupby('day', observed=True)['total_bill'].mean()
print("Средний total_bill по дням недели:")
for day, avg in avg_bill_by_day.items():
    print(f"   {day}: ${avg:.2f}")
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_bill = tips['total_bill'].median()
high_bill_tips = tips[tips['total_bill'] > median_bill]
avg_tip_by_sex = high_bill_tips.groupby('sex', observed=True)['tip'].mean()
print(f"Средний tip для счетов > медианы (${median_bill:.2f}) по полу:")
for sex, avg_tip in avg_tip_by_sex.items():
    print(f"   {sex}: ${avg_tip:.2f}")
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker_binary'] = tips['smoker'].map({'No': 0, 'Yes': 1})
print(tips[['smoker', 'smoker_binary']].head())
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.hist(tips['total_bill'], bins=20)
plt.title('Распределение total_bill')
plt.xlabel('Сумма счета ($)')
plt.ylabel('Частота')
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.figure(figsize=(10, 6))
plt.scatter(tips['total_bill'], tips['tip'], alpha=0.6)
plt.title('Взаимосвязь между total_bill и tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True)
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Распределение total_bill по дням недели')
plt.xlabel('День недели')
plt.ylabel('Total Bill ($)')
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.hist(tips[tips['time'] == 'Lunch']['tip'], bins=15, color='lightblue', edgecolor='black')
plt.title('Распределение tip (Lunch)')
plt.xlabel('Tip ($)')
plt.ylabel('Частота')


plt.subplot(1, 2, 2)
plt.hist(tips[tips['time'] == 'Dinner']['tip'], bins=15, color='lightcoral', edgecolor='black')
plt.title('Распределение tip (Dinner)')
plt.xlabel('Tip ($)')
plt.ylabel('Частота')
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

male_data = tips[tips['sex'] == 'Male']
scatter1 = ax1.scatter(male_data['total_bill'], male_data['tip'],
                      c=male_data['smoker'].map({'Yes': 'red', 'No': 'blue'}),
                      alpha=0.6)
ax1.set_title('Male: total_bill vs tip')
ax1.set_xlabel('Total Bill ($)')
ax1.set_ylabel('Tip ($)')
ax1.grid(True)

female_data = tips[tips['sex'] == 'Female']
scatter2 = ax2.scatter(female_data['total_bill'], female_data['tip'],
                      c=female_data['smoker'].map({'Yes': 'red', 'No': 'blue'}),
                      alpha=0.6)
ax2.set_title('Female: total_bill vs tip')
ax2.set_xlabel('Total Bill ($)')
ax2.set_ylabel('Tip ($)')
ax2.grid(True)

import matplotlib.patches as mpatches
red_patch = mpatches.Patch(color='red', label='Smoker')
blue_patch = mpatches.Patch(color='blue', label='Non-smoker')
plt.legend(handles=[red_patch, blue_patch], loc='upper center', bbox_to_anchor=(-0.1, -0.15), ncol=2)

plt.tight_layout()
plt.show()
```
